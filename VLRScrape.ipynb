```python
import datetime
import os
import pickle

TASKS_FILE = "tasks.pkl"

def display_reminder(task, due_date):
    """Display a reminder for a task."""
    print(f"Reminder: Task '{task}' is due on {due_date.strftime('%Y-%m-%d %H:%M')}")

def load_tasks():
    """Load tasks from a file."""
    if os.path.exists(TASKS_FILE):
        with open(TASKS_FILE, "rb") as f:
            return pickle.load(f)
    else:
        return {}

def save_tasks(tasks):
    """Save tasks to a file."""
    with open(TASKS_FILE, "wb") as f:
        pickle.dump(tasks, f)

def add_task(tasks):
    """Add a new task."""
    task_name = input("Enter task name: ")
    year = int(input("Enter year (YYYY): "))
    month = int(input("Enter month (MM): "))
    day = int(input("Enter day (DD): "))
    hour = int(input("Enter hour (24-hour format, HH): "))
    minute = int(input("Enter minute (MM): "))
    due_date = datetime.datetime(year, month, day, hour, minute)
    tasks[task_name] = due_date
    save_tasks(tasks)
    print("Task added successfully!")

def delete_task(tasks):
    """Delete an existing task."""
    task_name = input("Enter task name to delete: ")
    if task_name in tasks:
        del tasks[task_name]
        save_tasks(tasks)
        print("Task deleted successfully!")
    else:
        print("Task not found.")

def display_tasks(tasks):
    """Display all tasks."""
    if tasks:
        print("Tasks:")
        for task, due_date in tasks.items():
            print(f"- {task}: {due_date.strftime('%Y-%m-%d %H:%M')}")
    else:
        print("No tasks.")

def main():
    tasks = load_tasks()
    while True:
        print("\n1. Display tasks")
        print("2. Add task")
        print("3. Delete task")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            display_tasks(tasks)
        elif choice == "2":
            add_task(tasks)
        elif choice == "3":
            delete_task(tasks)
        elif choice == "4":
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

```python
import datetime
import os
import pickle

TASKS_FILE = "tasks.pkl"

def display_reminder(task, due_date):
    """Display a reminder for a task."""
    print(f"Reminder: Task '{task}' is due on {due_date.strftime('%Y-%m-%d %H:%M')}")

def load_tasks():
    """Load tasks from a file."""
    if os.path.exists(TASKS_FILE):
        with open(TASKS_FILE, "rb") as f:
            return pickle.load(f)
    else:
        return []

def save_tasks(tasks):
    """Save tasks to a file."""
    with open(TASKS_FILE, "wb") as f:
        pickle.dump(tasks, f)

def add_task(tasks):
    """Add a new task."""
    try:
        task_name = input("Enter task name: ")
        task_info = input("Enter year, month, day, hour, and minute separated by spaces: ")
        year, month, day, hour, minute = task_info.split()
        due_date = datetime.datetime(int(year), int(month), int(day), int(hour), int(minute))
        tasks.append({"name": task_name, "due_date": due_date})
        save_tasks(tasks)
        print("Task added successfully!")
        display_options(tasks)
    except ValueError:
        print("Invalid input. Please provide task name and task information correctly.")

def delete_task(tasks):
    """Delete an existing task."""
    try:
        task_number = int(input("Enter the number of the task to delete: "))
        if 1 <= task_number <= len(tasks):
            del tasks[task_number - 1]
            save_tasks(tasks)
            print("Task deleted successfully!")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Invalid input. Please enter a number.")
    display_options(tasks)

def display_tasks(tasks):
    """Display all tasks."""
    if tasks:
        print("Tasks:")
        for i, task_data in enumerate(tasks, start=1):
            print(f"{i}. {task_data['name']}: {task_data['due_date'].strftime('%Y-%m-%d %H:%M')}")
    else:
        print("No tasks.")
    display_options(tasks)

def display_options(tasks):
    """Display options after an action."""
    print("\n1. Display tasks")
    print("2. Add task")
    print("3. Delete task")
    print("4. Exit")

def main():
    tasks = load_tasks()
    display_options(tasks)
    while True:
        choice = input("Enter your choice: ")

        if choice == "1":
            display_tasks(tasks)
        elif choice == "2":
            add_task(tasks)
        elif choice == "3":
            delete_task(tasks)
        elif choice == "4":
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

```python
import requests
from bs4 import BeautifulSoup

# Send a GET request to the URL
url = "https://www.vlr.gg/stats/?event_group_id=61&event_id=all&region=all&country=all&min_rounds=200&min_rating=1550&agent=all&map_id=all&timespan=60d"
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table containing the data
table = soup.find('table')

# Extract headers from the table
headers = [header.text.strip() for header in table.find_all('th')]

# Extract data from the table
data = {}
for header in headers:
    data[header] = []

for row in table.find_all('tr'):
    cols = row.find_all('td')
    cols = [col.text.strip() for col in cols]
    if len(cols) == len(headers):  # Exclude header rows
        for i, header in enumerate(headers):
            data[header].append(cols[i])

# Convert numerical data to float for sorting
numerical_headers = ['K:D', 'ACS', 'ADR', 'KPR', 'APR']
for header in numerical_headers:
    data[header] = [float(value) if value else 0 for value in data[header]]

# Sort data by each numerical category and print top 5 players
for header in numerical_headers:
    print(f"Top 5 Players by {header}:")
    sorted_indices = sorted(range(len(data[header])), key=lambda i: data[header][i], reverse=True)[:5]
    for index in sorted_indices:
        player_name = data['Player'][index]
        formatted_player_name = ' '.join(player_name.split())  # Remove extra whitespaces
        print(f"{formatted_player_name}: {data[header][index]}")
    print()
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the page containing the select element
url = "https://www.vlr.gg/stats/?event_group_id=61&event_id=all&region=all&country=all&min_rounds=200&min_rating=1550&agent=all&map_id=all&timespan=60d"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find the select element by its name
select = soup.find("select", {"name": "event_id"})

# Check if the select element exists
if select:
    # Extract options
    options = select.find_all("option")

    # Initialize a list to store event IDs corresponding to Champions Tour 2024
    champions_tour_event_ids = []

    # Iterate over options and extract the event IDs corresponding to Champions Tour 2024
    for option in options:
        event_id = option["value"]
        name = option.text.strip()
        if "Champions Tour 2024" in name:
            champions_tour_event_ids.append(event_id)

    # Loop over each Champions Tour 2024 event ID and run the provided code
    for event_id in champions_tour_event_ids:
        # Send a GET request to the URL with the specific event ID
        event_url = f"https://www.vlr.gg/stats/?event_group_id=61&event_id={event_id}&region=all&country=all&min_rounds=200&min_rating=1550&agent=all&map_id=all&timespan=60d"
        event_response = requests.get(event_url)

        # Parse the HTML content
        event_soup = BeautifulSoup(event_response.text, 'html.parser')

        # Find the table containing the data
        table = event_soup.find('table')

        # Extract headers from the table
        headers = [header.text.strip() for header in table.find_all('th')]

        # Extract data from the table
        data = {}
        for header in headers:
            data[header] = []

        for row in table.find_all('tr'):
            cols = row.find_all('td')
            cols = [col.text.strip() for col in cols]
            if len(cols) == len(headers):  # Exclude header rows
                for i, header in enumerate(headers):
                    data[header].append(cols[i])

        # Convert numerical data to float for sorting
        numerical_headers = ['K:D', 'ACS', 'ADR', 'KPR', 'APR']
        for header in numerical_headers:
            data[header] = [float(value) if value else 0 for value in data[header]]

        # Sort data by each numerical category and print top 5 players
        print(f"Top 5 Players for Champions Tour 2024 Event ID: {event_id}")
        for header in numerical_headers:
            print(f"\nTop 5 Players by {header}:")
            sorted_indices = sorted(range(len(data[header])), key=lambda i: data[header][i], reverse=True)[:5]
            for index in sorted_indices:
                player_name = data['Player'][index]
                formatted_player_name = ' '.join(player_name.split())  # Remove extra whitespaces
                print(f"{formatted_player_name}: {data[header][index]}")
        print("\n--------------------------------------\n")

else:
    print("Select element not found.")
```
