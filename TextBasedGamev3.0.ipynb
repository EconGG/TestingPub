```python
from IPython.display import clear_output
import random
import time

exp = 0
base_hp = 30
level_up_modifier = 5

class Player:
  def __init__(self, name, hp=base_hp, exp=0):
    self.hp = hp
    self.name = name
    self.exp = exp
    self.base_damage_min = 1
    self.base_damage_max = 5

  def add_exp(self, amount):
   self.exp += amount
   print(f"Gained {amount} EXP! Total EXP: {self.exp}")

  def attack(self):
      total_damage = random.randint(self.base_damage_min, self.base_damage_max)
      return total_damage

  def set_base_damage_range(self, min_damage, max_damage):
    self.base_damage_min = min_damage
    self.base_damage_max = max_damage

class Bug:
    def __init__(self, hp):
        self.hp = hp

    def attack(self):
        return random.randint(1, 2)

#Create a Player Instance
player = Player("Hero")

def complete_quest(player, quest1_exp_reward):
  print("\nQuest completed!")
  player.add_exp(quest1_exp_reward)

def complete_battle(player, quest_exp_reward):
  print("\nBattle completed!")
  player.add_exp(quest_exp_reward)

def add_exp(current_exp, amount):
  current_exp += amount
  return current_exp

#STORYLINE & TEXT

def start_game():
  print("Welcome Traveller")
  player = Player("Hero", hp=base_hp)
  first_choice(player)

def get_choice(options, next_steps, player):
    while True:
        choice = input(f"\nEnter one of {options}: ")
        if choice in options:
            clear_output()
            next_steps[choice](player)
            break
        else:
            print(f"Invalid choice. Please enter one of {options}.")

def first_choice(player):
  print("\nYou find yourself at a crossroads. Do you:")
  print("1. Take the path through the dark forest")
  print("2. Head towards the village in the distance")

  options = ['1', '2']
  next_steps = {
      '1': dark_forest,
      '2': village
  }

  get_choice(options, next_steps, player)

def dark_forest(player):
  time.sleep(1)
  print("You enter the dark forest. It's quiet... too quiet.")
  print("Suddenly, you encounter a wild bug!")
  print("1. Fight the Bug!")
  print("2. Run Away")

  options = ['1', '2']
  next_steps = {
      '1': lambda player: battle(player, Bug(hp=15)),
      '2': keep_exploring
  }

  get_choice(options, next_steps, player)

# BATTLE

def battle(player, bug):
    time.sleep(1)
    clear_output()
    while player.hp > 0 and bug.hp > 0:
        print(f"\nPlayer HP: {player.hp} | Bug HP: {bug.hp}")

        # Player's turn
        print("\nWhat do you want to do?")
        print(f"1. Hit the bug. ({player.base_damage_min}-{player.base_damage_max})")
        print("2. Run away.")
        print("3. Fight until death.")
        choice = input("Enter your choice: ")

        if choice == '1':
            clear_output()
            player_damage = player.attack()
            bug.hp -= player_damage
            print(f"\nYou attack the bug and deal {player_damage} damage.")
            if bug.hp <= 0:
                print("You defeated the bug!")
                break
        elif choice == '2':
            clear_output()
            print("You ran away from the battle.")
            return
        elif choice == '3':
            while True:
                clear_output()
                player_damage = player.attack()
                bug.hp -= player_damage
                if bug.hp < 0:
                      bug.hp = 0
                else:
                  if bug.hp <= 0:
                    print(f"\nPlayer HP: {player.hp} | Bug HP: {bug.hp}")
                    print("You defeated the bug!")
                    break
                bug_damage = bug.attack()
                player.hp -= bug_damage
                print(f"The bug attacks you and deals {bug_damage} damage.")
                if player.hp <= 0:
                    print("You were defeated by the bug.")
                    return
        else:
            print("Invalid choice. Please enter 1, 2, or 3.")
            continue

        # Bug's turn
        if bug.hp > 0:
          bug_damage = bug.attack()
          player.hp -= bug_damage
          print(f"The bug attacks you and deals {bug_damage} damage.")
          if player.hp <= 0:
              print("You were defeated by the bug.")
              return
        else:
          break

    if player.hp > 0:
        bug_victory(player)
    else:
        game_over()

def bug_victory(player):
    time.sleep(1)
    bugbattle_exp_reward = 100
    complete_battle(player, bugbattle_exp_reward)

    print("\nWith the bug defeated, you find a treasure chest.")
    print("Do you:")
    print("1. Open the chest.")
    print("2. Leave it alone.")

    options = ['1', '2']
    next_steps = {
        '1': open_chest,
        '2': leave_chest
    }

    get_choice(options, next_steps, player)


def open_chest(player):
    time.sleep(1)
    print("\nYou open the chest and find a powerful sword!")
    print(f"\nDamage increase from [{player.base_damage_min}-{player.base_damage_max}] to [{player.base_damage_min+1}-{player.base_damage_max+1}]")
    player.set_base_damage_range(2,6)

    print("1. Keep Exploring")
    print("1. Go to the Village")

    options = ['1', '2']
    next_steps = {
      '1': keep_exploring,
      '2': village
  }

    get_choice(options, next_steps, player)

def leave_chest(player):
    print("\nYou decided not to open the chest.")
    print("1. Keep Exploring")
    print("1. Go to the Village")

    options = ['1', '2']
    next_steps = {
      '1': keep_exploring,
      '2': village
  }

    get_choice(options, next_steps, player)

#def weapon_increase(player):
#  print(f"\nDamage increase from [{player.base_damage_min}-{player.base_damage_max}] to [{player.base_damage_min+2}-{player.base_damage_max+1}]")
#  player.set_base_damage_range(4,7)
#  print(f"{player.base_damage_min}, {player.base_damage_max}")

def keep_exploring(player):
  time.sleep(1)
  print("\nYou don't find anything else in this forest.")

  print("1. Go to the Village")


  options = ['1']
  next_steps = {
      '1': village,
        }

  get_choice(options, next_steps, player)

def game_over():
  print("LOSE")

def forest_escape():
  print("You enter the Village")

def village(player):
  time.sleep(1)
  print("You enter the Village")

  print("1. Go to the Wizards Shack")
  print("2. Go to the Inn")
  print("3. Go to the Castle Doors.")

  options = ['1', '2', '3']
  next_steps = {
      '1': wizards_shack,
      '2': village,
      '3': village
        }

  get_choice(options, next_steps, player)

def wizards_shack(player):
  time.sleep(1)
  print("You enter the Wizard's shack and walk up to the counter.")
  print("\n The Wizard says aloud 'Greetings traveller, how can I help you'")

  print("1. Level Up")
  print("2. Shop")

  options = ['1', '2']
  next_steps = {
      '1': level,
      '2': village,
        }

  get_choice(options, next_steps, player)

def level(player):
  time.sleep(1)
  clear_output()
  print("The Wizard says 'If you have enough EXP, I can increase your max HP.")
  print(f"Total EXP: {player.exp}")
  print("\n1. Increase HP by 5")
  print("2. Look at other options")

  options = ['1', '2']
  next_steps = {
      '1': lambda player: level_up(player, base_hp, level_up_modifier),
      '2': wizards_shack,
        }

  get_choice(options, next_steps, player)

def level_up(player, base_hp, level_up_modifier):
  if player.exp >= 0:
    player.exp -= 100
    base_hp += level_up_modifier
    player.hp = base_hp
    print(f"HP Increased from {base_hp-level_up_modifier} -> {player.hp}")
    print(f"Current EXP: {player.exp}")
    level_up_modifier += 5
    print(f"Next time you level up you will gain: {level_up_modifier} HP")

start_game()
```
